The data is generated using sine function, therefore it is expected that as we increase `k`, it is easier to learn parabolics pattern of sine. # Sine Feature

```python
run_exp(train_path, sine=True, ks=[0, 1, 2, 3, 5, 10, 20])
plt.show()
```

By adding the sine function, the model is able to learn the parabolic curve very well, however there are noises that come from the coefficients of the polynomial terms. It is difficult to get coefficients exactly 0 in practice even though the actual model here is the simply $\sin(x) + \epsilon$.
